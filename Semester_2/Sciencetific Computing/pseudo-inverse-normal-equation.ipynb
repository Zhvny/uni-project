```python
# Import Libraries
import numpy as np

x = [1, 2, 3, 4, 5, 6, 7]
y = [3, 5, 6, 7, 9, 11, 13]

# Convert into numpy array
x  = np.array(x)
y = np.array(y)

# Persamaan Normal
# Change into matrix design
A = np.vstack((x, np.ones(len(x)))).T
alpha = np.dot(np.dot(np.linalg.inv(np.dot(A.T, A)), A.T), y)
print(f"y = {alpha[0].item():.2f}x + {alpha[1].item():.2f}")
```

```python
# Pseudo-invesre
A = np.vstack((x, np.ones(len(x)))).T
alpha = np.dot(np.linalg.pinv(A), y)
print(f"y = {alpha[0].item():.2f}x + {alpha[1].item():.2f}")

# Plotting
import matplotlib.pyplot as plt

plt.plot(x, y, 'b.')
# Plot hasil regression line
plt.plot(x, alpha[0]*x + alpha[1], 'r')
plt.xlabel("x")
plt.ylabel("y")
plt.legend(['x y coordinates', 'least square regression'])
plt.show()
```
